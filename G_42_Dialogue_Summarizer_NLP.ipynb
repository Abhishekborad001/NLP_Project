```python
import spacy
import re
# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Function to clean text and tokenize using spaCy
def clean_and_tokenize_spacy(text):
    if isinstance(text, list):
        text = ' '.join(text)

    # Clean the text
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = text.strip()

    # Tokenize with spaCy
    doc = nlp(text)
    return [token.text.lower() for token in doc]  # Return list of tokens in lowercase
!
# Apply text cleaning and tokenization on relevant columns
df['personas_tokens'] = df['personas'].apply(lambda x: clean_and_tokenize_spacy(x))
df['previous_utterance_tokens'] = df['previous_utterance'].apply(lambda x: clean_and_tokenize_spacy(x))
df['free_messages_tokens'] = df['free_messages'].apply(lambda x: clean_and_tokenize_spacy(x))

# Check the processed data
df[['personas', 'personas_tokens', 'previous_utterance', 'previous_utterance_tokens', 'free_messages', 'free_messages_tokens']].head()
```

```python
from rouge_score import rouge_scorer

# Initialize the ROUGE scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# Function to calculate ROUGE score for each summary
def calculate_rouge(reference_summary, generated_summary):
    scores = scorer.score(reference_summary, generated_summary)
    return scores

# Example of reference summaries (ground truths) – Replace with actual ground truth summaries
df['reference_summary'] = ["Your reference summary for this conversation"] * len(df)  # Replace with actual reference summaries

# Apply ROUGE score calculation
df['rouge_score'] = df.apply(lambda row: calculate_rouge(row['reference_summary'], row['summary']), axis=1)

# Check the results
print(df[['conversation_history', 'summary', 'rouge_score']].head())
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, BartTokenizer, BartForConditionalGeneration

# Load the pre-trained DistilBART model and tokenizer (for summarization)
tokenizer_bart = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
model_bart = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")

# Load the pre-trained DistilGPT-2 model and tokenizer (for response generation)
tokenizer_gpt2 = GPT2Tokenizer.from_pretrained("distilgpt2")
model_gpt2 = GPT2LMHeadModel.from_pretrained("distilgpt2")

# Set padding token to EOS token for DistilGPT-2
tokenizer_gpt2.pad_token = tokenizer_gpt2.eos_token

# Function to summarize the conversation using DistilBART
def summarize_conversation(text):
    inputs = tokenizer_bart(text, return_tensors="pt", max_length=512, truncation=True, padding=True)
    summary_ids = model_bart.generate(
        inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        max_length=50,
        num_beams=4,
        no_repeat_ngram_size=2,
        early_stopping=True
    )
    summary = tokenizer_bart.decode(summary_ids[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)
    return summary

# Function to generate a response using DistilGPT-2
def generate_response(text):
    inputs = tokenizer_gpt2(text, return_tensors="pt", max_length=512, truncation=True, padding=True)
    response_ids = model_gpt2.generate(
        inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        max_new_tokens=100,
        num_beams=5,
        no_repeat_ngram_size=2,
        early_stopping=True,
        pad_token_id=tokenizer_gpt2.eos_token_id
    )
    response = tokenizer_gpt2.decode(response_ids[0], skip_special_tokens=True)
    return response

# Chatbot interaction loop
conversation_history = ""

while True:
    # Get user input
    user_input = input("You: ")

    # If user wants to exit
    if user_input.lower() == 'exit':
        print("Bot: Goodbye!")
        break

    # Add user input to conversation history
    conversation_history += f"You: {user_input}\n"

    # Summarize the conversation history
    summarized_conversation = summarize_conversation(conversation_history)

    # Generate a response based on the summarized conversation
    bot_response = generate_response(summarized_conversation)

    # Display bot response
    print(f"Bot: {bot_response}")

    # Update conversation history with bot's response
    conversation_history += f"Bot: {bot_response}\n"
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, BartTokenizer, BartForConditionalGeneration
from datetime import datetime

# Load the pre-trained BART model and tokenizer (for summarization)
tokenizer_bart = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
model_bart = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")

# Load the pre-trained DialoGPT model and tokenizer (for response generation)
tokenizer_gpt2 = GPT2Tokenizer.from_pretrained("microsoft/DialoGPT-medium")
model_gpt2 = GPT2LMHeadModel.from_pretrained("microsoft/DialoGPT-medium")

# Set padding token to EOS token for DialoGPT
tokenizer_gpt2.pad_token = tokenizer_gpt2.eos_token

# Function for time-based dynamic greeting
def dynamic_greeting():
    current_hour = datetime.now().hour
    if current_hour < 12:
        return "Good morning! How can I assist you today?"
    elif current_hour < 18:
        return "Good afternoon! How can I assist you today?"
    else:
        return "Good evening! How can I assist you today?"

# Function to summarize the conversation using BART
def summarize_conversation(text):
    inputs = tokenizer_bart(text, return_tensors="pt", max_length=512, truncation=True, padding=True)
    summary_ids = model_bart.generate(
        inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        max_length=50,
        num_beams=4,
        no_repeat_ngram_size=2,
        early_stopping=True
    )
    summary = tokenizer_bart.decode(summary_ids[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)
    return summary

# Function to generate a response using DialoGPT
def generate_response(text):
    inputs = tokenizer_gpt2(text, return_tensors="pt", max_length=512, truncation=True, padding=True)
    response_ids = model_gpt2.generate(
        inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        max_new_tokens=100,
        num_beams=5,
        no_repeat_ngram_size=2,
        early_stopping=True,
        pad_token_id=tokenizer_gpt2.eos_token_id
    )
    response = tokenizer_gpt2.decode(response_ids[0], skip_special_tokens=True)
    return response

# Function to handle emotional cues (simple emotional response)
def handle_emotion(user_input):
    if "sad" in user_input.lower() or "upset" in user_input.lower():
        return "I'm really sorry to hear that. Do you want to talk about it?"
    elif "angry" in user_input.lower():
        return "I can understand you're upset. Let's try to work through it together."
    elif "happy" in user_input.lower():
        return "That's great to hear! What made your day better?"
    return ""

# Chatbot interaction loop
conversation_history = f"Bot: {dynamic_greeting()}\n"  # Initial greeting based on time of day

print(conversation_history)

while True:
    # Get user input
    user_input = input("You: ")

    # If user wants to exit
    if user_input.lower() in ['exit', 'bye', 'quit']:
        # Summarize the entire conversation before ending
        final_summary = summarize_conversation(conversation_history)
        print(f"Bot: Here's a summary of our conversation: {final_summary}")
        print("Bot: Goodbye! It was a pleasure chatting with you.")
        break

    # Handle emotional cues and provide empathetic responses
    emotional_response = handle_emotion(user_input)
    if emotional_response:
        print(f"Bot: {emotional_response}")
        conversation_history += f"Bot: {emotional_response}\n"

    # Add user input to conversation history
    conversation_history += f"You: {user_input}\n"

    # Summarize the conversation history
    summarized_conversation = summarize_conversation(conversation_history)

    # Generate a response based on the summarized conversation
    bot_response = generate_response(summarized_conversation)

    # Display bot response
    print(f"Bot: {bot_response}")

    # Update conversation history with bot's response
    conversation_history += f"Bot: {bot_response}\n"
```
